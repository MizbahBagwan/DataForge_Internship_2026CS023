```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("01_data/raw/dataset_raw.csv")

df.head()
```

```python
df.shape
```

```python
duplicates = df.duplicated().sum()

print("Duplicate Rows:", duplicates)
```

```python
df = df.drop_duplicates()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df["Order Year"] = df["Date"].dt.year

df["Order Month"] = df["Date"].dt.month_name()
```

```python
categorical = df.select_dtypes(include=["object", "string"]).columns

for col in categorical:
    df[col] = df[col].fillna("Unknown")
```

```python
numeric = df.select_dtypes(include="number").columns

for col in numeric:
    df[col] = df[col].fillna(df[col].median())
```

```python
print(df.columns.tolist())
```

```python
import numpy as np

df["High Value Order"] = np.where(
    df["Sales"] >= 1000,
    "Yes",
    "No"
)
```

```python
df["Sales Category"] = pd.cut(
    df["Sales"],
    bins=[0, 500, 1000, float("inf")],
    labels=["Low", "Medium", "High"]
)
```

```python
df["Total Price"] = df["Quantity"] * df["UnitPrice"]
```

```python
df.to_csv(
    "01_data/processed/dataforge_cleaned.csv",
    index=False
)

print("Dataset exported successfully!")
```
