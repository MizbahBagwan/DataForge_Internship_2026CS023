```python
schema = pd.DataFrame({
    "column": df.columns,
    "dtype": [str(df[c].dtype) for c in df.columns],
    "missing": [df[c].isna().sum() for c in df.columns],
    "unique": [df[c].nunique() for c in df.columns],
    "sample": [
        df[c].dropna().iloc[0] if not df[c].dropna().empty else ""
        for c in df.columns
    ]
})

schema
```

```python
schema.to_csv(
    "04_python/outputs/tables/schema_audit.csv",
    index=False
)

print("schema_audit.csv exported successfully!")
```
