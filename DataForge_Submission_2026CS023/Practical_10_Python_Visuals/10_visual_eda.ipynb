```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("01_data/processed/dataforge_cleaned.csv")
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Sales"], bins=20)
plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.savefig("04_python/outputs/plots/histogram_sales.png")
plt.show()
```

```python
plt.figure(figsize=(6,5))
plt.boxplot(df["Sales"])
plt.title("Sales Boxplot")
plt.savefig("04_python/outputs/plots/boxplot_sales.png")
plt.show()
```

```python
category_sales = df.groupby("Category")["Sales"].sum()

plt.figure(figsize=(8,5))
category_sales.plot(kind="bar")
plt.title("Sales by Category")
plt.savefig("04_python/outputs/plots/bar_category.png")
plt.show()
```

```python
year_sales = df.groupby("Order Year")["Sales"].sum()

plt.figure(figsize=(8,5))
plt.plot(year_sales.index, year_sales.values, marker="o")
plt.title("Sales Trend by Year")
plt.savefig("04_python/outputs/plots/line_sales.png")
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df["Quantity"], df["Sales"])
plt.title("Quantity vs Sales")
plt.xlabel("Quantity")
plt.ylabel("Sales")
plt.savefig("04_python/outputs/plots/scatter_quantity_sales.png")
plt.show()
```

```python
corr = df.select_dtypes(include="number").corr()

plt.figure(figsize=(8,6))
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap")
plt.savefig("04_python/outputs/plots/correlation_heatmap.png")
plt.show()
```

```python
missing = df.isna().sum()

plt.figure(figsize=(8,5))
missing.plot(kind="bar")
plt.title("Missing Values")
plt.savefig("04_python/outputs/plots/missing_values.png")
plt.show()
```

```python
region_sales = df.groupby("Region")["Sales"].sum()

plt.figure(figsize=(8,5))
region_sales.plot(kind="bar")
plt.title("Sales by Region")
plt.savefig("04_python/outputs/plots/region_sales.png")
plt.show()
```
