```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("01_data/processed/dataforge_cleaned.csv")
```

```python
df.head()
```

```python
numeric_summary = df.describe().T

numeric_summary
```

```python
numeric_summary.to_csv(
    "04_python/outputs/tables/numeric_summary.csv"
)
```

```python
missing = pd.DataFrame({
    "Missing Count": df.isna().sum(),
    "Missing Percent": (df.isna().mean()*100).round(2)
})

missing.sort_values(
    by="Missing Percent",
    ascending=False,
    inplace=True
)

missing
```

```python
missing.to_csv(
    "04_python/outputs/tables/missing_percent.csv"
)
```

```python
category_sales = (
    df.groupby("Category")["Sales"]
      .sum()
      .reset_index()
      .sort_values("Sales", ascending=False)
)

category_sales
```

```python
category_sales.to_csv(
    "04_python/outputs/tables/category_sales.csv",
    index=False
)
```

```python
region_sales = (
    df.groupby("Region")["Sales"]
      .sum()
      .reset_index()
      .sort_values("Sales", ascending=False)
)

region_sales
```

```python
region_sales.to_csv(
    "04_python/outputs/tables/region_sales.csv",
    index=False
)
```

```python
year_sales = (
    df.groupby("Order Year")["Sales"]
      .sum()
      .reset_index()
)

year_sales
```

```python
year_sales.to_csv(
    "04_python/outputs/tables/year_sales.csv",
    index=False
)
```

```python
category_quantity = (
    df.groupby("Category")["Quantity"]
      .sum()
      .reset_index()
      .sort_values("Quantity", ascending=False)
)

category_quantity
```

```python
category_quantity.to_csv(
    "04_python/outputs/tables/category_quantity.csv",
    index=False
)
```

```python
top10 = (
    df.nlargest(10, "Sales")
)

top10
```

```python
top10.to_csv(
    "04_python/outputs/tables/top10_sales.csv",
    index=False
)
```

```python
bottom10 = (
    df.nsmallest(10, "Sales")
)

bottom10
```

```python
bottom10.to_csv(
    "04_python/outputs/tables/bottom10_sales.csv",
    index=False
)
```
